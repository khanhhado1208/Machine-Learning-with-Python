## 3

```python
# Design matrix X = [1x10]
x = np.arange(1, 11)
x.reshape(1, 10)
```

```python
# Design matrix Y
y = np.array([14, 12, 34, 24, 43, 38, 41, 57, 55, 50])
y
```

```python
# a 10x2 matrix
np.vstack((x, y))
```

```python
# Regression coefficients
X = np.concatenate((np.ones((10,1)), x.reshape(10,1), x.reshape(10,1)**2), axis=1)
Y = y.reshape((10,1))
M = np.linalg.inv(np.dot(X.T, X)).dot(X.T).dot(Y)

xx = np.linspace(1, 10, 1000)
XX = np.concatenate((np.ones((1000,1)), xx.reshape(1000,1), xx.reshape(1000,1)**2), axis=1)

yy = XX.dot(M)
plt.plot(xx, yy)
plt.plot(x, y, 'o')
plt.show()
```

## 4

```python
def createGraph(polominal_degree):
    coef = np.polyfit(x.flatten(),y.flatten(), polominal_degree)
    p = np.poly1d(coef)
    xx = np.linspace (1,10,1000)
    yy = p(xx)
    plt.plot(x,y,'o', color= "blue")
    plt.plot(xx,yy, color= "black")
    plt.title(f"Polominal degree: {polominal_degree}")
    plt.show()
```

```python
for i in range(0,13):
    createGraph(i)
```

# Conclusion:
